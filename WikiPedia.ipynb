# My Favorite Things All-In-One Code

Get A summary on all search results </br />
an enter into a SQLite database.<br />
Thsi also generates a text file with the name of the search -spaces<br />
EXAMPLE: a search on ----  Origin of water on Earth <br />
results in database entry AND   Wiki_OriginofwateronEarth.txt

```python
# Wiki Search -Summary of search results print to text
# AND to a database
import sqlite3
from time import sleep
from shutil import copyfile
import wikipedia
#backup your current datafile incase of a problem
dbname = 'WIKIData.db'
conn = sqlite3.connect(dbname) 
copyfile('WIKIData.db','WIKIData.db-bak')
topic = raw_input("Research : ")
fname = topic.replace(" ", "")+".txt"
fname = "Wiki_"+fname

conn.text_factory = str
c = conn.cursor()
c.execute('''
CREATE VIRTUAL TABLE IF NOT EXISTS wiki
USING FTS3(row, para);
''')
f =open(fname, "w")
f.close()

rows = wikipedia.search(topic)
for row in rows:
    row = row.encode('utf-8').strip()
    row = str(row)
    para = wikipedia.summary(row)
    para =  para.encode('utf-8').strip() 
    para = '\n\n'.join((row, para))
    enter = open(fname, "a")
    c.execute("INSERT INTO wiki VALUES (?,?)", (row, para))
    conn.commit()
    enter.write(para)
    enter.close()
    print para
enter.close()
```
